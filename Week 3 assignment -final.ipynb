# The first part of the assignment

```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
```

```python
filename ="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
dfs = pd.read_html(filename)
df=dfs[0]
df.columns = ['Postal code', 'Borough', 'Neighborhood']
df.head()
```

```python
# Ignore cells with a borough that is Not assigned
df=df[df['Borough']!= 'Not assigned']
df.head(15)
```

```python
# combine neighborhoods that exist in one postal code area

df.loc[df['Neighborhood'] =='Not assigned' , 'Neighborhood'] = df['Borough']
```

```python
# use the neighborhood to replace a borough with Not assigned value
df.Neighborhood.replace("Not assigned", df.Borough, inplace=True)
df.head(15)
```

```python
for index, row in df.iterrows():
    if row["Neighborhood"] == "Not assigned":
        row["Neighborhood"] = row["Borough"]
```

Suppose we are shopping in a Mall and want to find a Chinese restaurant nearby to have lunch. The following work is used for locating all the Chinese restuarants around and helps us to know the ratings

```python
# Choose a random mall in downtown Toronto: Dragon City Mall

address = '280 Spadina Ave, Toronto, ON'
geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

Search for a specific venue category：Chinese restaurant

```python
search_query = 'Chinese restaurant'
radius = 300
print(search_query + ' .... OK!')
```

```python
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query, radius, LIMIT)
```

```python
results = requests.get(url).json()
results
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']
```

```python
# tranform venues into a dataframe
df2 = pd.json_normalize(venues)
df2.head(40)
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in df2.columns if col.startswith('location.')] + ['id']
df2_filtered = df2.loc[:, filtered_columns]
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# filter the category for each row
df2_filtered['categories'] = df2_filtered.apply(get_category_type,axis=1)
# clean column names by keeping only last term
df2_filtered.columns = [column.split('.')[-1] for column in df2_filtered.columns]
df2_filtered
```

```python
# keep only the Chinese Restaurants in catagories
df3=df2[df2_filtered['categories']=='Chinese Restaurant']
df3
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=13) #generate map 
# add a red circle marker to represent the Dragon City Mall
folium.features.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='Dragon City Mall',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.3
).add_to(venues_map)
```

Explore the rating of a Given Venue

```python
# the function to get the ratings of the restaurants by using their IDs
def ratings(venue_id):
    url = 'https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}'.format(venue_id, CLIENT_ID,CLIENT_SECRET, VERSION)
    result = requests.get(url).json()
    result['response']['venue']
    try:
        print(result['response']['venue']['rating'])
    except:
        print('This venue has not been rated yet.')
```
